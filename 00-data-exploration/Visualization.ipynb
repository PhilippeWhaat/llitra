```python
datos = datos.fillna(value=0)
```

```python
datos = datos.loc[:, ~datos.eq(0).all()]
datos.columns
```

```python
visu = datos.groupby("Tipo_de_documento").sum()
visu = visu.drop(columns=['Termino', 'Folio', 'Unnamed: 0.1'])
#visu.reset_index(level=0, inplace=True)
visu.columns
```

```python
visu = visu.loc[:, ~datos.eq(0).all()]
```

```python
pivote = pd.pivot(visu, columns=['apellido_otro', 'apellido_autoridad', 'direccion_otro', 'direccion_actor', 'direccion_demandado', 'fecha_matrimonio', 'firma_otro', 'firma_demandado', 'monto_gastos', 'nombre_otro', 'nombre_actor', 'nombre_autoridad', 'nombre_autorizado', 'nombre_benef_test', 'nombre_demandado', 'num_cuenta_demandado', 'num_expediente', 'num_juzgado', 'ID_acta_matrim'])
pivote
```

```python
fig, axs = plt.subplots()

# Una vez que creamos eso, podemos pegar la gráfica con un argumento adicional
# en lo que acabamos de crear, con "ax"

# eso permite manipular el marco si no tenemos el argumento figsize
fig, axs = plt.subplots(figsize=(25,6))

sns.heatmap(visu, cmap="Blues", ax=axs)
plt.show()
```
